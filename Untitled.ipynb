```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import math
import numpy as np

# Configuration
INPUT_CSV = "features.csv"
TARGET_CELL_INDEX = 0  # We will visualize the distribution for Cell 0 across all images
SUBPLOTS_PER_ROW = 4   # Constraint: Max 4 subplots per row

def visualize_feature_distribution(df, feature_names, group_name):
    """
    Plots Histogram and Boxplot for a list of features.
    Layout: 2 Features per row (Feature A Hist, Feature A Box | Feature B Hist, Feature B Box)
    """
    num_features = len(feature_names)
    if num_features == 0:
        print(f"No features found for {group_name}")
        return

    # We fit 2 features per row (because each feature gets 2 plots: Hist + Box)
    # Total subplots per row = 4
    rows = math.ceil(num_features / 2)
    
    # Dynamic Figure Size
    fig, axes = plt.subplots(rows, SUBPLOTS_PER_ROW, figsize=(20, 5 * rows))
    fig.suptitle(f"Distribution Analysis: {group_name} (Cell Index {TARGET_CELL_INDEX})", fontsize=16)
    
    # Flatten axes array for easy indexing
    axes = axes.flatten()
    
    for i, feature in enumerate(feature_names):
        # Calculate subplot indices
        # Feature 0 goes to index 0 (Hist) and 1 (Box)
        # Feature 1 goes to index 2 (Hist) and 3 (Box)
        ax_hist_idx = i * 2
        ax_box_idx = i * 2 + 1
        
        if ax_hist_idx >= len(axes):
            break

        # Extract data (drop NaNs for plotting)
        data = df[feature].dropna()

        # 1. Histogram (Distribution)
        sns.histplot(data=data, kde=True, ax=axes[ax_hist_idx], color='skyblue')
        axes[ax_hist_idx].set_title(f"{feature} - Histogram")
        axes[ax_hist_idx].set_xlabel("Value")
        axes[ax_hist_idx].set_ylabel("Frequency")

        # 2. Box Plot (Outliers)
        sns.boxplot(x=data, ax=axes[ax_box_idx], color='lightgreen')
        axes[ax_box_idx].set_title(f"{feature} - Boxplot")
        axes[ax_box_idx].set_xlabel("Value")

    # Turn off any unused subplots (if you have an odd number of features)
    total_plots_needed = num_features * 2
    for j in range(total_plots_needed, len(axes)):
        axes[j].axis('off')

    plt.tight_layout(rect=[0, 0.03, 1, 0.95]) # Adjust for suptitle
    plt.show()

def main():
    print("Loading data...")
    try:
        df = pd.read_csv(INPUT_CSV)
    except FileNotFoundError:
        print(f"Error: {INPUT_CSV} not found.")
        return

    # 1. Filter for ONE Cell
    print(f"Filtering data for Cell Index: {TARGET_CELL_INDEX}")
    df_cell = df[df['CellIndex'] == TARGET_CELL_INDEX]
    
    if df_cell.empty:
        print("No data found for this cell index.")
        return

    print(f"Analyzing {len(df_cell)} images for Cell {TARGET_CELL_INDEX}...")

    # --- Define Groups ---
    all_cols = df_cell.columns.tolist()
    
    # Group 1: Shape Features (Only 2 features)
    shape_feats = ['Shape_Lines', 'Shape_Circles']
    
    # Group 2: Convolution Features (First 8 for brevity, or all 48 if you prefer)
    # Showing first 8 to keep plot readable
    conv_feats = [c for c in all_cols if "Conv" in c][:8] 
    
    # Group 3: Color Features (First 8 - e.g., Red channel bins)
    color_feats = [c for c in all_cols if "Color" in c][:8]
    
    # Group 4: HOG Features (First 8 only - there are 3000+!)
    hog_feats = [c for c in all_cols if "HOG" in c][:8]

    # --- Generate Plots ---
    
    # Plot Shape
    print("Plotting Shape Features...")
    visualize_feature_distribution(df_cell, shape_feats, "Shape Features")
    
    # Plot Convolution
    print("Plotting Convolution Features (Sample)...")
    visualize_feature_distribution(df_cell, conv_feats, "Convolution Features")
    
    # Plot Color
    print("Plotting Color Features (Sample)...")
    visualize_feature_distribution(df_cell, color_feats, "Color Features")

    # Plot HOG
    print("Plotting HOG Features (Sample)...")
    visualize_feature_distribution(df_cell, hog_feats, "HOG Features (First 8)")

if __name__ == "__main__":
    main()
```

```python
import pandas as pd
import numpy as np
from statsmodels.stats.outliers_influence import variance_inflation_factor
from statsmodels.tools.tools import add_constant
import warnings

warnings.filterwarnings("ignore")

# Configuration
INPUT_CSV = "features.csv"
OUTPUT_CSV = "features_vif_reduced.csv"
VIF_THRESHOLD = 10
# Sampling is CRITICAL for VIF speed. We calculate VIF on a sample, 
# determine which columns to keep, then apply to the full dataset.
SAMPLE_SIZE = 5000 

def reduce_dimensions():
    print("1. Loading Data...")
    df = pd.read_csv(INPUT_CSV)
    
    # Separate Metadata (Don't calculate VIF on these)
    metadata_cols = ['ImageName', 'CellIndex', 'Label']
    df_meta = df[metadata_cols]
    df_features = df.drop(columns=metadata_cols)
    
    print(f"Original Feature Count: {df_features.shape[1]}")

    # --- PRE-PROCESSING ---
    # 1. Handle NaNs and Infinite values
    df_features = df_features.replace([np.inf, -np.inf], np.nan).fillna(0)

    # 2. Remove Constant Columns (Variance = 0)
    # VIF crashes on constant columns
    var_check = df_features.var()
    drop_constants = var_check[var_check == 0].index
    df_features = df_features.drop(columns=drop_constants)
    print(f"Dropped {len(drop_constants)} constant columns.")

    # --- OPTIMIZATION: SAMPLING ---
    # If dataset is huge, VIF will take days. We sample to find collinearity structure.
    if len(df_features) > SAMPLE_SIZE:
        print(f"Data too large for rapid VIF. Sampling {SAMPLE_SIZE} rows for calculation...")
        df_sample = df_features.sample(n=SAMPLE_SIZE, random_state=42)
    else:
        df_sample = df_features

    # --- STAGE 1: CORRELATION FILTER (Pre-VIF) ---
    # Remove features with correlation > 0.95 first. 
    # This prevents 'Singular Matrix' errors and speeds up VIF by 100x.
    print("2. Stage 1: Removing Highly Correlated Features (> 0.95)...")
    corr_matrix = df_sample.corr().abs()
    upper = corr_matrix.where(np.triu(np.ones(corr_matrix.shape), k=1).astype(bool))
    to_drop_corr = [column for column in upper.columns if any(upper[column] > 0.95)]
    
    df_sample = df_sample.drop(columns=to_drop_corr)
    print(f"Dropped {len(to_drop_corr)} features due to high correlation.")
    print(f"Remaining features for VIF: {df_sample.shape[1]}")

    # --- STAGE 2: ITERATIVE VIF ---
    print(f"3. Stage 2: Iterative VIF Removal (Threshold={VIF_THRESHOLD})...")
    print("This may take some time. Please wait.")
    
    # We loop until all features have VIF < Threshold
    while True:
        # Add constant (required for VIF calculation in statsmodels)
        X_const = add_constant(df_sample)
        
        # Calculate VIF for all current features
        vif_data = pd.Series(
            [variance_inflation_factor(X_const.values, i) for i in range(X_const.shape[1])],
            index=X_const.columns
        )
        
        # Drop the 'const' index derived from add_constant
        if 'const' in vif_data.index:
            vif_data = vif_data.drop('const')
        
        # Get Max VIF
        max_vif = vif_data.max()
        max_feature = vif_data.idxmax()
        
        if max_vif > VIF_THRESHOLD:
            # Drop the worst offender
            df_sample = df_sample.drop(columns=[max_feature])
            print(f"   Dropping '{max_feature}' (VIF={max_vif:.2f})")
        else:
            print(f"   Converged! Max VIF is {max_vif:.2f}")
            break

    # --- FINALIZATION ---
    # Apply the columns kept in df_sample to the FULL original dataset
    cols_to_keep = df_sample.columns.tolist()
    
    print("4. Applying reduction to full dataset...")
    df_final = pd.concat([df_meta, df_features[cols_to_keep]], axis=1)
    
    print(f"Final Shape: {df_final.shape}")
    print(f"Saving to {OUTPUT_CSV}...")
    df_final.to_csv(OUTPUT_CSV, index=False)
    print("Done.")

if __name__ == "__main__":
    reduce_dimensions()
```

```python
import pandas as pd
import numpy as np
import warnings

# Suppress warnings
warnings.filterwarnings("ignore")

# Configuration
INPUT_CSV = "features.csv"
OUTPUT_CSV = "features_vif_optimized.csv"
VIF_THRESHOLD = 10
CORRELATION_THRESHOLD = 0.90  # More aggressive initial filter to ensure matrix invertibility

def fast_vif_reduction():
    print("1. Loading Data...")
    df = pd.read_csv(INPUT_CSV)
    
    # Separate Metadata
    metadata_cols = ['ImageName', 'CellIndex', 'Label']
    df_meta = df[metadata_cols]
    df_features = df.drop(columns=metadata_cols)
    
    print(f"Original Feature Shape: {df_features.shape}")

    # --- PRE-PROCESSING ---
    # Fill NaNs and drop constants (Variance=0)
    df_features = df_features.replace([np.inf, -np.inf], np.nan).fillna(0)
    var_check = df_features.var()
    drop_constants = var_check[var_check == 0].index
    df_features = df_features.drop(columns=drop_constants)
    print(f"Dropped {len(drop_constants)} constant columns.")

    # --- STAGE 1: Aggressive Correlation Filter ---
    # We calculate the Correlation Matrix ONCE.
    # This is O(N * M^2) but we only do it one time.
    print("2. Calculating Correlation Matrix...")
    corr_matrix = df_features.corr()
    
    # Filter highly correlated features (> 0.90)
    # This is critical to prevent "Singular Matrix" errors in the inversion step
    print(f"   Filtering features with correlation > {CORRELATION_THRESHOLD}...")
    
    # Get upper triangle of correlation matrix
    upper = corr_matrix.where(np.triu(np.ones(corr_matrix.shape), k=1).astype(bool))
    
    # Find columns to drop
    to_drop = [column for column in upper.columns if any(upper[column].abs() > CORRELATION_THRESHOLD)]
    
    # Drop them from our tracking list and the matrix
    kept_features = [c for c in df_features.columns if c not in to_drop]
    # Resize the correlation matrix to match kept features
    corr_matrix = corr_matrix.loc[kept_features, kept_features]
    
    print(f"   Dropped {len(to_drop)} features via correlation filter.")
    print(f"   Remaining features: {len(kept_features)}")

    # --- STAGE 2: Matrix Inversion VIF (The Fast Part) ---
    print(f"3. Iterative VIF Removal (Threshold={VIF_THRESHOLD})...")
    
    dropped_by_vif = []
    
    # Loop until max VIF is below threshold
    while True:
        # Check for singularity (Infinite VIF)
        try:
            # Invert Correlation Matrix: O(M^3)
            # This is much faster than running M linear regressions
            inv_corr = np.linalg.inv(corr_matrix.values)
        except np.linalg.LinAlgError:
            print("   Warning: Matrix Singular. Dropping feature with highest correlation sum...")
            # Heuristic: Drop feature that correlates most with everything else
            sum_corr = corr_matrix.abs().sum().sort_values(ascending=False)
            worst_feat = sum_corr.index[0]
            print(f"   Dropping {worst_feat} (Singular Fix)")
            
            kept_features.remove(worst_feat)
            corr_matrix = corr_matrix.drop(index=worst_feat, columns=worst_feat)
            continue

        # VIF is the diagonal of the inverse correlation matrix
        vifs = np.diag(inv_corr)
        
        # Find Max VIF
        max_vif = vifs.max()
        max_idx = vifs.argmax()
        max_feature = kept_features[max_idx]
        
        if max_vif > VIF_THRESHOLD:
            # print(f"   Dropping '{max_feature}' (VIF={max_vif:.2f})") # Uncomment for verbose
            if len(kept_features) % 100 == 0:
                print(f"   ... {len(kept_features)} features remaining (Current Max VIF: {max_vif:.2f})")
                
            # Remove from list
            kept_features.pop(max_idx)
            dropped_by_vif.append(max_feature)
            
            # Efficiently slice the correlation matrix (pandas is optimized for this)
            # We DO NOT re-calculate correlations from the raw data.
            corr_matrix = corr_matrix.drop(index=max_feature, columns=max_feature)
        else:
            print(f"   Converged! Max VIF is {max_vif:.2f}")
            break

    print(f"   Dropped {len(dropped_by_vif)} features via VIF.")

    # --- SAVE ---
    print("4. Saving reduced dataset...")
    df_final = pd.concat([df_meta, df_features[kept_features]], axis=1)
    df_final.to_csv(OUTPUT_CSV, index=False)
    print(f"Done. Final dimensions: {df_final.shape}")

if __name__ == "__main__":
    fast_vif_reduction()
```

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.decomposition import PCA
from sklearn.manifold import TSNE
from sklearn.preprocessing import StandardScaler
import warnings

warnings.filterwarnings("ignore")

# Configuration
INPUT_CSV = "features.csv" # Or "features_vif_optimized.csv"
SEED = 42

def plot_tsne_variance():
    print("1. Loading Data...")
    try:
        df = pd.read_csv(INPUT_CSV)
    except FileNotFoundError:
        print(f"Error: {INPUT_CSV} not found.")
        return

    # Separate Metadata
    # We only want to run math on the Feature columns
    metadata_cols = ['ImageName', 'CellIndex', 'Label']
    
    # Check if metadata columns exist, otherwise guess
    available_meta = [c for c in metadata_cols if c in df.columns]
    
    y = df['Label'] if 'Label' in df.columns else None
    X = df.drop(columns=available_meta, errors='ignore')
    
    # Handle NaNs/Infs
    X = X.replace([np.inf, -np.inf], np.nan).fillna(0)
    
    print(f"   Original Dimensions: {X.shape[1]} features")

    # --- STEP 1: SCALING (Crucial) ---
    print("2. Standardizing Data...")
    scaler = StandardScaler()
    X_scaled = scaler.fit_transform(X)

    # --- STEP 2: PCA (95% Variance) ---
    print("3. Applying PCA (Keeping 95% Variance)...")
    # n_components=0.95 tells PCA to choose the minimum number of 
    # components needed to explain 95% of the variance.
    pca = PCA(n_components=0.95, random_state=SEED)
    X_pca = pca.fit_transform(X_scaled)
    
    components_kept = X_pca.shape[1]
    print(f"   PCA Reduced dimensions from {X.shape[1]} to {components_kept}")
    print(f"   (These {components_kept} components explain 95% of the information)")

    # --- STEP 3: t-SNE ---
    print("4. Running t-SNE on PCA output...")
    # We feed the PCA result (X_pca) into t-SNE, not the raw data.
    # This removes noise and makes t-SNE run 10x faster.
    tsne = TSNE(n_components=2, perplexity=30, random_state=SEED, n_jobs=-1)
    X_tsne = tsne.fit_transform(X_pca)

    # --- STEP 4: VISUALIZATION ---
    print("5. Plotting...")
    
    # Combine t-SNE results with Labels for plotting
    plot_df = pd.DataFrame(X_tsne, columns=['t-SNE Dimension 1', 't-SNE Dimension 2'])
    if y is not None:
        plot_df['Label'] = y

    plt.figure(figsize=(12, 8))
    
    sns.scatterplot(
        x='t-SNE Dimension 1', 
        y='t-SNE Dimension 2',
        hue='Label',
        palette='viridis',
        data=plot_df,
        alpha=0.7,
        s=60 # Dot size
    )
    
    plt.title(f"t-SNE Projection of Top 95% Variance\n(Reduced from {X.shape[1]} features to {components_kept} PCA components)", fontsize=14)
    plt.grid(True, linestyle='--', alpha=0.3)
    plt.legend(bbox_to_anchor=(1.05, 1), loc='upper left')
    plt.tight_layout()
    plt.show()

if __name__ == "__main__":
    plot_tsne_variance()
```
